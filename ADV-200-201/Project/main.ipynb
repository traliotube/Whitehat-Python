```python
#start of predefine code
from ipywidgets import widgets
import IPython.display
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from ipyfilechooser import FileChooser


graphType = ["Choose one.. ","bubble","bar"]
functionalityList = ["Choose One","Sort","Filter"]
sortOption = ["Ascending","Descending"]
df = ""
newDf = ""
inputBox = ""
inputFontsize = ""
inputTitle = ""

def selectFiles(b):
    IPython.display.clear_output()
    global df
    global graphType
    df = pd.read_csv(fc.selected)
    print(fc.selected)
    df.replace( "", float("nan"), inplace=True)
    df.replace( "", float("nan"), inplace=True)
    df.dropna(inplace=True)
    IPython.display.display(df)
    functionalityWidget = widgets.Dropdown(options = functionalityList)
    functionalityInt = widgets.interactive(functionality, functionalityOpt = functionalityWidget)
    IPython.display.display(functionalityInt)

def functionality(functionalityOpt):
    global sortOption
    global df
    global inputBox
    if functionalityOpt == "Sort":
        sortColWidget = widgets.Dropdown(options = df.columns)
        sortOptWidget = widgets.Dropdown(options = sortOption)
        rangeWidget = widgets.Dropdown(options = [5, 10, 20, 30, 40, 50])
        sortInt = widgets.interactive(sortDf, sortCol=sortColWidget, sortOpt = sortOptWidget, range=rangeWidget)
        IPython.display.display(sortInt)
    elif functionalityOpt == "Filter":
        inputBox = widgets.Text(description="Value:")
        IPython.display.display(inputBox)
        filterColWidget = widgets.Dropdown(options = df.columns)
        compareOptWidget = widgets.Dropdown(options = ["Choose the option..","=", ">","<"])
        rangeWidget = widgets.Dropdown(options = [5, 10, 20, 30, 40, 50])
        groupByInt = widgets.interactive(filterDf, filterCol=filterColWidget, range=rangeWidget, compareOpt = compareOptWidget)
        IPython.display.display(groupByInt)
    else:
        print("Choose a function")

def sortDf(sortCol, sortOpt, range):
    global df
    global newDf
    print(df[sortCol].dtypes)
    if df[sortCol].dtypes != "float64" or "int64":
        df[sortCol] = df[sortCol].astype("float64")
    if sortOpt == "Ascending":
        newDf = df.sort_values(by=[sortCol], ascending=True)
    elif sortOpt == "Descending":
        newDf = df.sort_values(by=[sortCol], ascending=False)
    newDf = newDf.head(range)
    IPython.display.display(newDf)
    getWidget()

def filterDf(filterCol, range, compareOpt):
    global df
    global newDf
    global inputBox
    # try:
    if compareOpt == "=":
        newDf = df.loc[df[filterCol] == float(inputBox.value)]
    elif compareOpt == ">":
        if df[filterCol].dtypes != 'float64' or 'int64':
            df[filterCol] = df[filterCol].astype('float64')
        newDf = df.loc[df[filterCol] > float(inputBox.value)]
    elif compareOpt == "<":
        if df[filterCol].dtypes != 'float64' or 'int64':
            df[filterCol] = df[filterCol].astype('float64')
        newDf = df.loc[df[filterCol] < float(inputBox.value)]
    else: 
        print("Choose a Valid Operation")
    newDf = newDf.head(range)
    IPython.display.display(newDf)
    getWidget()
    # except:
    #     print("The data is not structured so cannot perform the selected action")

def getWidget():
    global df
    global newDf
    global graphType
    global inputTitle
    global inputFontsize
    xlabelWidget = widgets.Dropdown(options = newDf.columns)   
    ylabelWidget = widgets.Dropdown(options = newDf.columns)
    inputTitle = widgets.Text(description="Title of Graph:")
    inputFontsize = widgets.Text(description="Font Size:")
    IPython.display.display(inputTitle)
    IPython.display.display(inputFontsize)
    graphWidget = widgets.Dropdown(options = graphType)
    graph = widgets.interactive(displayPlot, xaxis=xlabelWidget, yaxis= ylabelWidget, graphType = graphWidget)
    IPython.display.display(graph)         

def displayPlot(xaxis, yaxis, graphType):
    global newDf
    global inputTitle
    global inputFontsize
    plt.subplots(figsize=(19,8))
    if(graphType == "bubble"):
        rgb = np.random.rand(3)
        plt.scatter(newDf[xaxis], newDf[yaxis], c=rgb, alpha=1, s=newDf[yaxis]*10)
        plt.title(inputTitle.value, fontsize=inputFontsize.value)
        plt.xlabel(xaxis, fontsize=inputFontsize.value)
        plt.xticks(rotation="vertical")
        plt.ylabel(yaxis, fontsize=inputFontsize.value)
        plt.show()
    elif(graphType == "bar"):
        plt.bar(newDf[xaxis], newDf[yaxis], color=["red", "green","blue","yellow","pink"])
        plt.title(inputTitle.value, fontsize=inputFontsize.value)
        plt.xlabel(xaxis, fontsize=inputFontsize.value)
        plt.xticks(rotation="vertical")
        plt.ylabel(yaxis, fontsize=inputFontsize.value)
        plt.show()
    else:
        print("Choose valid graph")


fc = FileChooser()
fc.register_callback(selectFiles)
IPython.display.display(fc)
#end of predefined
```
