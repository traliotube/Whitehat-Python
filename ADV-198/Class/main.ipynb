```python
#import libraries and csv
import pandas as pd
import matplotlib.pyplot as plt 

pd.set_option('display.max_columns', None)

df = pd.read_csv('Energy Census and Economic Data US 2010-2014.csv')
df

```

```python
#Get the column number of POPESTIMATE2010 
POPESTIMATE2010 = df.columns.get_loc("POPESTIMATE2010")
POPESTIMATE2010
```

```python
#Get the column number of POPESTIMATE2014
POPESTIMATE2014 = df.columns.get_loc("POPESTIMATE2014")
POPESTIMATE2014
```

```python
#Create new dataframe only for Population colums and state column
populationDf = df[df.columns[POPESTIMATE2010:POPESTIMATE2014+1] ] 
populationDf['State'] = df['State']
populationDf

```

```python

#from population_dataframe find the row for New Jersey and create a new dataframe
newJerseyPopulation= populationDf.loc[populationDf['State'] == 'New Jersey']
newJerseyPopulation
 
```

```python
#Get the column number of NatGasC2010 
NatGasC2010 = df.columns.get_loc("NatGasC2010")
NatGasC2010
```

```python
#Get the column number of NatGasC2014
NatGasC2014 = df.columns.get_loc("NatGasC2014")
NatGasC2014
```

```python
#Create new dataframe only for Natural Gas colums
naturalGasDataframe = df[df.columns[NatGasC2010:NatGasC2014 + 1] ] 
naturalGasDataframe['State'] = df['State']
naturalGasDataframe

```

```python
#from natural_gas_dataframe find the row for New Jersey and create a new dataframe
newJerseyGas= naturalGasDataframe.loc[naturalGasDataframe['State'] == 'New Jersey']
newJerseyGas
 
```

```python
#Multiply each natural gas year value with 10 for making it normalize
newJerseyGas['NatGasC2010'] = newJerseyGas['NatGasC2010']*10
newJerseyGas['NatGasC2011'] = newJerseyGas['NatGasC2011']*10
newJerseyGas['NatGasC2012'] = newJerseyGas['NatGasC2012']*10
newJerseyGas['NatGasC2013'] = newJerseyGas['NatGasC2013']*10
newJerseyGas['NatGasC2014'] = newJerseyGas['NatGasC2014']*10
newJerseyGas
```

```python
#Create a common dataframe for Population and NatGasConsumption
columnsGas = ['Year' , 'Population','NatGasConsumption']
indexGas  = [1,2,3,4,5]
dataGas =[
    [2010,newJerseyPopulation['POPESTIMATE2010'].values[0],newJerseyGas['NatGasC2010'].values[0]],
    [2011,newJerseyPopulation['POPESTIMATE2011'].values[0],newJerseyGas['NatGasC2011'].values[0]],
    [2012,newJerseyPopulation['POPESTIMATE2012'].values[0],newJerseyGas['NatGasC2012'].values[0]],
    [2013,newJerseyPopulation['POPESTIMATE2013'].values[0],newJerseyGas['NatGasC2013'].values[0]],
    [2014,newJerseyPopulation['POPESTIMATE2014'].values[0],newJerseyGas['NatGasC2014'].values[0]],

]
finalGas = pd.DataFrame(dataGas, index=indexGas, columns=columnsGas)
finalGas

```

```python
#plot a line graph for showing Natural Gas Consumption vs Population
fig = plt.subplots(figsize=(10,6))
label = finalGas['Year']
value = finalGas['NatGasConsumption']


plt.plot(label, value, label = "NatGasConsumption" , linewidth=3.0)

label = finalGas['Year']
value = finalGas['Population']
plt.plot(label, value, label = "Population" , linewidth=3.0)

plt.xlabel('Population')

plt.ylabel('NatGasConsumption')

plt.title('NatGasConsumption VS Population in New Jersey', fontsize=20)

plt.legend()

plt.show()
```

Conslusion : The Natural Gas Consumption and Population growth was constant from 2010 till 2012 in New Jersey, after 2012 there was significant growth in Population which lead to significant growth Natural Gas Consumption

# Activity 2 - Find the Correlation between the Coal Consumption and Coal Price  from 2010-2014.

```python
#predefine code 
#import libraries and csv
import pandas as pd
import matplotlib.pyplot as plt 

pd.set_option('display.max_columns', None)

df = pd.read_csv('Energy Census and Economic Data US 2010-2014.csv')
df

#predefine code end
```

```python
#Get the column number of CoalC2010 
CoalC2010 = df.columns.get_loc("CoalC2010")
CoalC2010
```

```python
#Get the column number of CoalC2014 
CoalC2014 = df.columns.get_loc("CoalC2014")
CoalC2014
```

```python
#Create new dataframe only columns for Coal
coal = df[df.columns[CoalC2010:CoalC2014+1]]
coal["State"] = df["State"]
coal
```

```python
#from Coal_dataframe find the row for New Mexico and create a new dataframe
coal = coal.loc[coal["State"] == "New Mexico"]
coal
```

```python
#Get the column number of CoalPrice2010
CoalPrice2010 = df.columns.get_loc("CoalPrice2010")
CoalPrice2010
```

```python
#Get the column number of CoalPrice2014
CoalPrice2014 = df.columns.get_loc("CoalPrice2014")
CoalPrice2014
```

```python
#Create new dataframe only  columns for Coal_price
coalPrice = df[df.columns[CoalPrice2010:CoalPrice2014+1]]
coalPrice["State"] = df["State"]
coalPrice
```

```python
#from Coal_price_dataframe find the row for New Mexico and create a new dataframe
coalPrice = coalPrice.loc[coalPrice["State"] == "New Mexico"]
coalPrice
```

```python
#Multiply each CoalPrice year wise column values with 100000 for making it normalize
for i in coalPrice:
    if i != "State":
        coalPrice[i] *= 100000
coalPrice
```

```python
#Create a common dataframe for coalConsumption and coalPrice
columnsCoal = ['Year' ,'coalConsumption', 'coalPrice']
indexCoal  = [1,2,3,4,5]
dataCoal =[
    [2010,coal['CoalC2010'].values[0],coalPrice['CoalPrice2010'].values[0]],
    [2011,coal['CoalC2011'].values[0],coalPrice['CoalPrice2011'].values[0]],
    [2012,coal['CoalC2012'].values[0],coalPrice['CoalPrice2012'].values[0]],
    [2013,coal['CoalC2013'].values[0],coalPrice['CoalPrice2013'].values[0]],
    [2014,coal['CoalC2014'].values[0],coalPrice['CoalPrice2014'].values[0]],

]
finalCoal = pd.DataFrame(dataCoal, index=indexCoal, columns=columnsCoal)
finalCoal
```

```python
#Plot a line graph showing the correlation between the Coal Consumption and Coal Price 

#predefine code end
fig = plt.subplots(figsize=(19,8))


#update label and value variable
label = finalCoal["Year"]
value = finalCoal["coalConsumption"]
plt.plot(label, value, label = "Coal Consumption" , linewidth=3.0)

#update label and value variable
label = finalCoal["Year"]
value = finalCoal["coalPrice"]
plt.plot(label, value, label = "Coal Price" , linewidth=3.0)


plt.xlabel('Dates')
plt.ylabel('Coal Consumption VS Coal Price')
plt.title('Coal Consumption VS Coal Price', fontsize=20)
plt.legend()
plt.show()

#predefine code end
```
