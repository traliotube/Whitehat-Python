```python
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
```

```python
#Task 1 
#Sort the data as per total number of cases
t2 = df.sort_values(by="total_cases")
t2
```

```python
#Task 2
#Get top 5 countries who has the least number of cases and plot a bar graph
t2 = t2.head(5)
plt.xlabel("Country")
plt.xticks(rotation='vertical')
plt.ylabel("Total Cases")
label = t2["country"]
value = t2["total_cases"]
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue'))
```

```python
#Task 3
#Sort the data as per total number of deaths
t3 = df.sort_values('total_deaths', ascending=False)
t3
```

```python
#Task 4
#Get top 5 countries who has the maximum number of deaths and plot a bar graph
t3 = t3.head(5)
plt.xlabel("Country")
plt.xticks(rotation='vertical')
plt.ylabel("Total Deaths")
label = t3["country"]
value = t3["total_deaths"]
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue'))
```

```python
#Task 5
#Sort the data as per active cases
t4 = df.sort_values('active_cases', ascending=False)
t4
```

```python
#Task 6
#Get top 5 countries who has the maximum number of active cases and plot a bar graph
t4 = t4.head(5)
plt.xlabel("Country")
plt.xticks(rotation='vertical')
plt.ylabel("Active Cases")
label = t4["country"]
value = t4["active_cases"]
plt.bar(label,value,width=0.4, color=('red','yellow','green','pink','blue'))
```
