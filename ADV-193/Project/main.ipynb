```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('cardiovascular.csv')
df
```

# Average cholestrol found in various age groups

```python
#Task 2 
#Group by age and find the average cholesterol and make a dataframe out of it
age = df.groupby("age")["cholesterol"].mean().reset_index(name="cholesterol")
age
```

```python
#Task 3 
#Plot a line graph for various age group and their cholesterol 
label = age['age']
value = age['cholesterol']
fig = plt.subplots(figsize=(19,8))
plt.plot(label, value, label = "Average cholestrol found in various age groups" , linewidth=3.0)
plt.xlabel('Age')
plt.xticks(rotation='vertical')
plt.ylabel('Cholesterol')
plt.title('Average cholestrol found in various age groups', fontsize=20)

plt.legend()

plt.show()
```

# Correlation between systolic and diastolic blood pressure

```python
# Diastolic blood pressure - is the pressure in the arteries when the heart rests between beats
# Systolic blood pressure - the pressure in your arteries when your heart beats

#predefine code for image
from IPython.display import Image
Image(filename='img.jpg') 
#predefine code end
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it
systolic = df.groupby("age")["systolic_blood_pressure"].max().reset_index(name="systolic")
systolic
```

```python
#Task 5
#Group by age and find maximum diastolic blood pressure and create a dataframe project out of it
diastolic = df.groupby("age")["diastolic_blood_pressure"].max().reset_index(name="diastolic")
diastolic
```

```python
#Task 6
#Plot a line graph to show a Correlation between systolic and diastolic blood pressure
fig = plt.subplots(figsize=(19,8))

plt.plot(diastolic['age'], diastolic['diastolic'], label = "Diastolic blood pressure" , linewidth=3.0)
plt.plot(systolic['age'], systolic['systolic'], label = "Systolic blood pressure" , linewidth=3.0)

plt.xlabel('Age')
plt.xticks(rotation='vertical')

plt.title('Correlation between systolic and diastolic blood pressure', fontsize=20)

plt.legend()

plt.show()
```
