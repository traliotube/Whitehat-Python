# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#read the csv
df = pd.read_csv("framingham.csv")
df
```

```python
#Filter and make a new dataframe for non smokers
nonSmokers = df.loc[df["currentSmoker"] == 0]
nonSmokers
```

```python
#Group by age column and find average heart rate at different age
nonSmokers = nonSmokers.groupby("age")["heartRate"].mean().reset_index()
nonSmokers
```

```python
#Filter and make a new dataframe for smokers
smokers = df.loc[df["currentSmoker"] == 1]
smokers
```

```python
#Group by age column and find average heart rate at different age
smokers = smokers.groupby("age")["heartRate"].mean().reset_index()
smokers
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers
fig = plt.subplots(figsize=(19,8))
plt.plot(nonSmokers["age"], nonSmokers["heartRate"], label = "Non Smokers" , linewidth=3.0)
plt.plot(smokers["age"], smokers["heartRate"], label = "Smokers" , linewidth=3.0)

plt.title("Difference between heart rate of smokers and non smokers")
plt.xlabel('Age')
plt.ylabel('Heart Rate')
plt.legend()
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv

#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
t2 = df.loc[df["TenYearCHD"] == 1]
t2
```

```python
#Group by age column and count the rows of TenYearCHD column
t2 = t2.groupby('age')["TenYearCHD"].count().reset_index()
t2
```

```python
#Plot a bubble graph to show total number of people having a chance of coronary heart disease in 10 years
plt.figure(figsize=(12, 8))
plt.scatter(t2['age'], t2['TenYearCHD'], 
                  c=[np.random.rand(3)],label = 'Chance of coronary heart disease in 10 years', 
                 alpha=0.5,
                 s = t2['TenYearCHD']*25)

plt.legend()
plt.xlabel("Age", size=14)
plt.ylabel("Chance", size=14)
```
