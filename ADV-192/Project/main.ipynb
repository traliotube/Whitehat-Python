```python
#BMI Data

#predefine code
import pandas as pd
import matplotlib .pyplot as plt

dataframe = pd.read_csv("bmi.csv")
df = dataframe.dropna()
bmi = df['bmi']
df
```

```python
#Task 1
#How many people are underweight and create a dataframe out of it
under = df.loc[df["bmi"]<18.9]["gender"].reset_index(name="gender")
underCount = under["index"].count()
print(underCount)
under
```

```python
#Task 2
#How many people have normal weight and create a dataframe out of it
normal = df.loc[df["bmi"]>18.9]["gender"].reset_index(name="gender")
normal = df.loc[df["bmi"]<24.5]["gender"].reset_index(name="gender")
normalCount = normal["index"].count()
print(normalCount)
normal
```

```python
#Task 3
#Plot a pie chart as per the percentage of people who are underweight and healthy. 
value = [underCount, normalCount]
label = ["Under Weight", "Normal Weight"]
plt.pie(value, labels=label, autopct="%0.1f%%")
```

```python
#Task 4
#Group by the gender from underweight dataframe and create another data frame out of it
genderUnder = under.groupby("gender")["gender"].count().reset_index(name="number")
genderUnder
```

```python
#Task 5
#Plot a pie chart as per the percentage of male and female who are underweight
plt.pie(genderUnder["number"], labels=genderUnder["gender"], autopct="%0.1f%%")
```

```python
#Task 6
#Group by the gender from healthy weight dataframe and create another data frame out of it
genderNormal = normal.groupby("gender")["gender"].count().reset_index(name="number")
genderNormal
```

```python
#Task 7
#Plot a pie chart as per the percentage of male and female who are healthy
plt.pie(genderNormal["number"], labels=genderNormal["gender"], autopct="%0.1f%%")
```
